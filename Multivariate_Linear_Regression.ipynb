#  Multivariate Linear Regression

## Multiple Features
Linear regression that uses multiple variables is known as "*multivariate linear regression*".

```python
import tensorflow as tf
import numpy as np

theta_values = tf.constant([2.0, 3.0, 4.0], shape=[3,1])
x_values = tf.constant([1.0,3.0,9.0,], shape=[3,1])
result = tf.linalg.matmul(theta_values,x_values,transpose_a=True)
result
```

## Gradient Decent

### Feature Scaling
In certain situations the gradient decent algorithm will diverge on a long path when the contour map is not scaled properly. In order to fix this we will scale the features to have the same range.

### Mean Normalization
To make the features have approximately zero mean we replace xi with (xi-ui)/s1.

### Learning Rate
To determine if our gradient decent is working correctyl we can plot J(theta) vs number of iterations. 
If the graph does not look like its converging to 0 then it is a clear indication that you must use a 
smaller alpha value.
- **Note:** If alpha is sufficiently small then it should decrease on every iteration.

**In theory**
- If alpha is too small - slow convergence
- If alpha is too large - may not decrease/converge
