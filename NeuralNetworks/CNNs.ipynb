# Convolutional Neural Networks

A **Convolutional Neural Netowrk** popularly known as _ConvNet_ or _CNN_ is a deep learning algorithm that can take an image as an input and learn the weights and biases to various aspects/objects in that image.

The inspiration for CNNs came from the connectivity pattern seen in the visual cortex of the human brain. Each neuron responds to stimuli only in a restricted region of the visual field known as the receptive field. A collection of such fields overlap to cover the entire visual area.

## Why use CNNs?
CNNs are better at capturing the spatial and temporal dependencies in an image. This architecture peroms a better fitting due to reduction in the number of parameters involved and reusability of weights.

The role of the ConvNet is to reduce the images into a form which is easier to process, without losing features which are critical for getting a good prediction.

## Convolutional Layer -- The Kernel
The element involved in carrying out the convolution operation in the first part is known as the **kernel/filter**. The kernel is a smaller matrix that performs **matrix multiplication operation** by striding through the matrix until it parses the entire width and height.


### Kernel objective
The objective of the convolution operation is to **extract the high-level features** from the input image. These layers build upon the previous infrences. For example layer one may find small edges. Second layer may put theses edges together to form shapes. Then more complex features and so on.
