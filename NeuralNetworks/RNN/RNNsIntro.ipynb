# Recurrent Neural Networks

## What are RNNs?
Unlike regular perceptrons or convolutional networks, recurrent networks have a small memory loop. In a more formal definition, _"They are a class of neural networks that allow previous outputs to be used as inputs while having hidden states."_ [(CS 230)](https://stanford.edu/~shervine/teaching/cs-230/cheatsheet-recurrent-neural-networks)

## Advantages and Drawbacks
The main speciality in RNNs is the ability to model short term dependencies. it retains information from one time step to another flowing through the unrolled RNN units.

Advantages:
   - Ability to process an input of any length
   - Size of the model not increasing with input
   - Computation taking into account of previous information
   - Weights are shared across time

Drawbacks:
   - Computationally slow
   - Difficult to access information from a long time ago
   - Can't consider future inout for the current state

## Applications
RRN models are mostly used in the fields of natural processing and speech recognition. 
Some of the examples are:
   - Music generation
   - Sentiment classification
   - Name entity recognition
   - Machine translation

## Architecture
There are a lot of differnt variations in which a RNN can be created.

Here is a list of applications for each of the differnt variations formed using RNNs:
1. One to One
    - This architecture works just like a traditional neural network. There's nothing special about it.
2. One to Many
    - Popularly usef for image captioning and music generation.
3. Many to One
    - Used for sentiment analysis
4. Many to Many (type 1)
    - Used in named entity recognition
5. Many to Many (type 2)
    - Used for machine translation from one language to another.
